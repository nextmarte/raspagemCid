```python
# Importar bibliotecas necessárias
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np

# URL do site
url = "https://maricainfo.com/itinerarios-dos-onibus-vermelhinhos-marica"

# Baixar dados do site
response = requests.get(url)

# Criar objeto BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Extrair texto
text = soup.get_text()

# Remover parágrafos e quebras de linha
text = text.replace('\n', '')

# Remover todos os elementos antes de E01 e todos depois de PRAÇA DO FERREIRINHA
text = re.search('E01.*PRAÇA DO FERREIRINHA', text).group()

# Quebrar uma linha sempre que encontrar o padrão regex LINHA E10 ou LINHA E10A
text = re.sub('E\\d+[A-Z]?', '\n\\g<0>', text)

# Dividir o texto em um vetor de linhas
lines = text.split('\n')

# Criar um DataFrame com as linhas de ônibus
df = pd.DataFrame(lines, columns=['Linha'])

# Visualizar o DataFrame
df.tail(10)
```

```python
# Extrair o número da linha e criar uma nova coluna 'NumeroLinha'
df['NumeroLinha'] = df['Linha'].str.extract('(E\d+[A-Z]?)')
df['Linha'] = df['Linha'].apply(lambda x: re.sub('(E\d+[A-Z]?\s)', '', x))
# Visualizar o DataFrame


df.head(36)
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_site():
    """
    Scrapes the given website and prints the href attribute of all anchor tags.

    Returns:
    None
    """
    url = "https://maricainfo.com"

    response = requests.get(url)

    soup = BeautifulSoup(response.text, 'html.parser')

    for link in soup.find_all('a'):
        print(link.get('href'))

scrape_site()
```
